### Ilk adim: Basit Bir Yapi <a name="1"></a>
x ve y adinda iki adet noktalar iceren dizimiz olsun. Bu noktalarin kesismelerini birlestirerek bir grafik ortaya cikaralim.

````python
  x = [1, 5, 8]
  y = [1, 3, 11]
  plt.plot(x, y)
  plt.title("Tek Dogru Grafigi") #Grafik Basligi
  plt.xlabel("x") #x dogrusu ismi
  plt.ylabel("y") #y dogrusu ismi
  plt.legend(["x ve y dogrusu"]) #Grafikteki cizginin ismi
  plt.show() #Grafigi goster.
````

### Daha Fazla noktasal kume olursa? <a name="2"></a>


````python
  x = [1, 5, 8]
  y = [1, 3, 11]
  z = [8, 4, 1]
  t = [0, 7, 4]
  plt.plot(x, y)
  plt.plot(x, z)
  plt.plot(x, t)
  plt.title("Iki Noktasal Kume Grafigi")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.legend(["x ve y", "x ve z", "x ve t"]) 
  plt.show()
````

## Ilk Uygulama: Ne Kadar Genc! <a name="4"></a>
**Simdi uygulama zamani** <br/>
FIFA 19 dataset'i ile islemlere baslamistik simdi ise ilk uygulamaya basliyoruz. <br/>
#### UYGULAMA 
Secilen 3 kulubun yas ortalamalarini alalim ve bunlari bir grafikte gosterelim.

````python
#Kulüp isimlerini dataframe'den çekiyoruz.
def club(club_Name):
    return data[data['Club']==club_Name][['Age']]
````

````python
club_List = ['Juventus', 'Real Madrid', 'Roma']
club_Mean = []
for n in club_List:
    x=club(n)
    a = x['Age'].mean()
    club_Mean.append(a)
````

````python
fig, ax = plt.subplots(figsize=(10, 8))
ax.bar(club_List, club_Mean)
ax.set(xlabel='Age Mean', ylabel='Club Name', title='Uygulama 1')

#plt.figure(figsize = (10, 8))
#plt.bar(club_List, club_Mean)
#plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10,8))
ax.barh(club_List, club_Mean)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlabel='Age Mean', ylabel='Club Name', title='Uygulama 1')
````
